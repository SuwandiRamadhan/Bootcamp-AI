# Data yang sudah di cleaning AMES HOUSING

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as pl
```

```python
data_ames_housing = pd.read_csv('./data_baru_bersih.csv')
data_ames_housing
```

```python
data_ames_housing.info()
```

```python
# Fill data kosong di kolom electrical
data_ames_housing['Electrical'] = data_ames_housing['Electrical'].fillna('No Electrical')

# Fill data kosong di kolom Lot Frontage
data_ames_housing['Lot Frontage'] = data_ames_housing['Lot Frontage'].fillna(data_ames_housing['Lot Frontage'].mean())

# Fill data kosong di kolom Mas Vnr Type
data_ames_housing['Mas Vnr Type'] = data_ames_housing['Mas Vnr Type'].fillna('No Mas Vnr')

# Fill data kosong di kolom BsmtFin SF 1 
data_ames_housing['BsmtFin SF 1'] = data_ames_housing['BsmtFin SF 1'].fillna(data_ames_housing['BsmtFin SF 1'].mean())

# Fill data kosong di kolom BsmtFin SF 2
data_ames_housing['BsmtFin SF 2'] = data_ames_housing['BsmtFin SF 2'].fillna(data_ames_housing['BsmtFin SF 2'].mean())

# Fill data kosong di kolom Bsmt Unf SF
data_ames_housing['Bsmt Unf SF'] = data_ames_housing['Bsmt Unf SF'].fillna(data_ames_housing['Bsmt Unf SF'].mean())

# Fill data kosong di kolom Total Bsmt SF
data_ames_housing['Total Bsmt SF'] = data_ames_housing['Total Bsmt SF'].fillna(data_ames_housing['Total Bsmt SF'].mean())

# Fill data kosong di kolom Bsmt Full Bath
data_ames_housing['Bsmt Full Bath'] = data_ames_housing['Bsmt Full Bath'].fillna(data_ames_housing['Bsmt Full Bath'].mean())

# Fill data kosong di kolom Bsmt Half Bath
data_ames_housing['Bsmt Half Bath'] = data_ames_housing['Bsmt Half Bath'].fillna(data_ames_housing['Bsmt Half Bath'].mean())

# Fill data kosong di kolom Garage Cars
data_ames_housing['Garage Cars'] = data_ames_housing['Garage Cars'].fillna(data_ames_housing['Garage Cars'].mean())

# Fill data kosong di kolom Garage Area
data_ames_housing['Garage Area'] = data_ames_housing['Garage Area'].fillna(data_ames_housing['Garage Area'].mean())
```
